# Practicas 

>**En este notbook lo que se hace es tomar un dataframe de un archivo trials.csv que contiene informacio del enjuiciamiento de brujas desde 1300 a 1800 y apartir de este abrir la informacion en mas tablas que luegos seran insertadas dentro de una base de datos mysql mediante pandas y su funcion sql**

```python
# Si no estan instaladas instalar con el comando pip install
import pandas as pd # pip install pandas
import sqlalchemy as db # pip install sqlalchemy
import mysql.connector as conector # pip install mysql-connector-python
```

Traemos el csv y visualizamos el dataframe

```python
dfWitchs = pd.read_csv('trials.csv')
dfWitchs
```

Realizo conexión con base de datos MySql con el conctor de mysql y creo cursor

```python
# Conectamos con la base de datos
conn = conector.connect(
    host="localhost",
    user="root",
    password=""
)

cursor = conn.cursor(buffered=True)
```

#### Creamos base de datos y tablas en el servidor

```python
createDatabaseQuery = "CREATE DATABASE IF NOT EXISTS witch"
cursor.execute(createDatabaseQuery)
```

```python
createCityTableQuery = "CREATE TABLE IF NOT EXISTS witch.city (`id` INT(11) NOT NULL AUTO_INCREMENT, `city` VARCHAR(45) NULL DEFAULT NULL, PRIMARY KEY (`id`))"
cursor.execute(createCityTableQuery)
```

```python
createCenturyTableQuery = "CREATE TABLE IF NOT EXISTS witch.century (`id` INT(11) NOT NULL AUTO_INCREMENT,`century` INT(5) NULL DEFAULT NULL,PRIMARY KEY (`id`))"
cursor.execute(createCenturyTableQuery)
```

```python
createDecadeTableQuery = "CREATE TABLE IF NOT EXISTS witch.decade (`id` INT(11) NOT NULL AUTO_INCREMENT,`decade` INT(5) NULL DEFAULT NULL,PRIMARY KEY (`id`))"
cursor.execute(createDecadeTableQuery)
```

```python
createDeathsTriedTableQuery = """
                CREATE TABLE IF NOT EXISTS witch.death_tried (
                `id` INT(11) NOT NULL AUTO_INCREMENT,
                `deaths` INT(10) NULL DEFAULT NULL,
                `tried` INT(10) NULL DEFAULT NULL,
                `city_id` INT(11) NOT NULL,
                `century_id` INT(11) NOT NULL,
                `decade_id` INT(11) NOT NULL,
                PRIMARY KEY (`id`),                
                CONSTRAINT `fk_death_tried_city1`
                    FOREIGN KEY (`city_id`)
                    REFERENCES witch.`city` (`id`)
                    ON DELETE NO ACTION
                    ON UPDATE NO ACTION,
                CONSTRAINT `fk_death_tried_century1`
                    FOREIGN KEY (`century_id`)
                    REFERENCES witch.`century` (`id`)
                    ON DELETE NO ACTION
                    ON UPDATE NO ACTION,
                CONSTRAINT `fk_death_tried_decade1`
                    FOREIGN KEY (`decade_id`)
                    REFERENCES witch.`decade` (`id`)
                    ON DELETE NO ACTION
                    ON UPDATE NO ACTION)
"""
cursor.execute(createDeathsTriedTableQuery)
```

Realizamos el commit y cerramos la conexión con el cursor

```python
conn.commit()
```

```python
cursor.close()
```

#### Conectamos mediante sqlalchemy y su engine al servidor mysql

```python
engine = db.create_engine("mysql+pymysql://root:@localhost/witch")
connection = engine.connect()
```

```python
dfWitchs
```

Realizamos la extraccion de la informacion del dataset de cada una de las tablas que vamos a utilizar, luego la insertamos en la base de datos en su tabla correspondiente y seguido verificamos mediante una consulta su inserción en la base de datos

```python
dfDecade = dfWitchs[['decade']]
dfDecade
```

```python
dfDecade.to_sql(name='decade', con=connection, if_exists='append', index = False)
```

```python
q = "SELECT * FROM decade"
dataDecade = pd.read_sql(q , con=connection)
dataDecade
```

```python
dfCentury = dfWitchs[['century']]
dfCentury
```

```python
dfCentury.to_sql(name='century', con=connection, if_exists='append', index = False)
```

```python
q = "SELECT * FROM century"
dataCentury = pd.read_sql(q , con=connection)
dataCentury
```

```python
dfCity = dfWitchs[['city']]
dfCity
```

```python
dfCity.to_sql(name='city', con=connection, if_exists='append', index = False)
```

```python
q = "SELECT * FROM city"
dataCity = pd.read_sql(q , con=connection)
dataCity
```

Para el caso de la tabla de muerte y enjuiciamientos tenemos que hay que colocar las claves foraneas para esto simplemente hace falta crear un dataframe con las nuevas columnas que seran las claves foraneas y seguidamente insertar los datos

```python
dfDeathsTried = dfWitchs[['deaths','tried']]
dfDeathsTried['city_id'] = dataCity['id']
dfDeathsTried['century_id'] = dataCentury['id']
dfDeathsTried['decade_id'] = dataDecade['id']
dfDeathsTried
```

```python
dfDeathsTried.to_sql(name='death_tried', con=connection, if_exists='append', index = False)
```

```python
q = "SELECT * FROM death_tried"
dataDeathsTried = pd.read_sql(q , con=connection)
dataDeathsTried
```
