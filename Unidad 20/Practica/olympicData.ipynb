### Practica Dataframe y bases de datos

```python
import sqlite3
import pandas as pd
import sqlalchemy as db
```

1. Conectarse a la url que contiene el archivo CSV de medallas olímpicas.

```python
url = 'http://winterolympicsmedals.com/medals.csv'
dfMedals = pd.read_csv(url)
dfMedals.sample(5)
```

2. Descargar los datos y obtener un sub dataset que contenga a todas las medallas de oro (Gold) Estados Unidos (USA) a partir del año 1950.

```python
dfFiltered = dfMedals[(dfMedals['Medal'] == 'Gold') & (dfFiltered['NOC'] == 'USA') & (dfFiltered['Year'] >= 1950)]
dfFiltered
```

3. Crear una base de datos “olympics” en SQLite y la tabla “medals”.

```python
# Crear una conexión a la base de datos
conn = sqlite3.connect('olympics.db')

# Crear un cursor
cursor = conn.cursor()

# Crear una tabla
query = """CREATE TABLE IF NOT EXISTS medals (
            id INTEGER,
            year INTEGER,
            city TEXT,
            sport TEXT,
            discipline TEXT,
            noc TEXT,
            event TEXT,
            event_gender TEXT,
            medal TEXT
            )"""
cursor.execute(query)
conn.commit()
conn.close()
```

4. Guardar en la base de datos. los datos generados en el sub dataset.

```python
# Creación del engine
engine = db.create_engine('sqlite:///olympics.db')
connection = engine.connect()

# Insertar datos en la BD utilzando el método to_sql
dfFiltered.to_sql(name='orders', con=connection, if_exists='append', index = False)
```

5. Consultar la base de datos y validar que los datos se hayan cargado correctamente

```python
# Verificamos que todo se haya incertado correctamente
query = "SELECT * FROM orders"
data = pd.read_sql(query, con=connection)
data
```
