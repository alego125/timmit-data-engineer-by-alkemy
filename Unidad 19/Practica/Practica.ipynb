**6) Realizar las siguientes acciones sobre el segundo dataframe:**
  * **Group By sobre un campo y utilizando sum()**
  * **Realizar un Melt**

```python
# Juicios por brujeria por siglo
triedByCentury = dfCsv.groupby(['century'])[['tried']].sum()
triedByCentury.reset_index(inplace=True)
triedByCentury
```

```python
dfToMelt = dfExcel[['Year','Entity','Outdoor air pollution']].dropna(axis=0, how="any", thresh=None, subset=None, inplace=False)
dfToMelt
```

```python
# Traspuesta
dfToMelt.melt()
```

```python
dfExcel
```

```python
# Muertes por abuso de alcohol por año
dfAlcoholUse = dfExcel.groupby(['Year'])[['Alcohol use']].sum()
dfAlcoholUse.reset_index(inplace=True)
dfAlcoholUse
```

```python
dfDeaths = dfCsv.groupby(['city'], as_index=False)[['deaths','tried']].sum()
dfDeaths = dfDeaths[dfDeaths['deaths'] > 100 ]
dfDeaths
```

**7) Realizar un gráfico de Barras con el primer dataframe**

```python
triedByCentury.plot(x='century', y='tried', kind = 'line')

plt.show()
```

```python
dfAlcoholUse.plot(x='Year', y='Alcohol use', kind = 'bar')

plt.show()
```

```python
plt.bar(dfDeaths['city'],dfDeaths['deaths'], rotation=90)
plt.bar(dfDeaths['city'],dfDeaths['tried'], rotation=90)
plt.show()
```
